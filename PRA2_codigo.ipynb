# Pregunta 1

**¿Tener una relación puede afectar a la percepción que se tiene del uso de internet?**

```python
# Excluimos registros con NaN en la opinión sobre internet

df1 = df[ df["P16"] < 6 ]
```

```python
rangos_edades = sorted(list(set(df["EDAD_RANGO"])))
```

```python
opinion_media_16_soltero = []
opinion_media_16_pareja = []
```

```python
for ages in rangos_edades:
    
    p = list( df1[ ( df1["EDAD_RANGO"]==ages ) & ( df1["PAREJA"]=='Con pareja') ]["P16"])
    s = list( df1[ ( df1["EDAD_RANGO"]==ages ) & ( df1["PAREJA"]=='Soltero/a') ]["P16"])
    
    opinion_media_16_pareja.append(std.mean(p))
    opinion_media_16_soltero.append(std.mean(s))
```

```python
# Creamos el dataframe de opinion sobre uso de internet según relación

df1_opinion = pd.DataFrame()

df1_opinion["edad"] = rangos_edades
df1_opinion["solteros"] = opinion_media_16_soltero
df1_opinion["con pareja"] = opinion_media_16_pareja

df1_opinion
```

# Pregunta 2

**¿La satisfacción en la vida familiar afecta a la satisfacción en distintos campos relacionados a la vida social o afectiva?**

```python
# Excluimos outliers

df2 = df[ (df["P1_1"]<15) & (df["P1_2"]<15) & (df["P1_3"]<15) ]
```

```python
indices = list(range(0,len(df2)))
```

```python
# Barajamos lista de índices

rd.shuffle(indices)
```

```python
len(indices)
```

```python
re = []

for i in range(1,len(indices)):
    
    if len(indices)%i == 0:
        re.append(i)

print(re)
```

```python
# Agruparemos 44 grupos con 53 muestras aleatorias cada uno

2332/44
```

```python
rel_familia = []
rel_social = []
rel_afectiva = []

i = 5

while ( i < len(indices) ):
    
    j = 0
    aux_f = []
    aux_s = []
    aux_a = []
    
    while ( j < 53 ):
        
        aux_f.append(list(df2["P1_1"])[indices[0]])
        aux_s.append(list(df2["P1_2"])[indices[0]])
        aux_a.append(list(df2["P1_3"])[indices[0]])
        
        indices.pop(0)
        j = j + 1
        
    rel_familia.append(std.mean(aux_f))
    rel_social.append(std.mean(aux_s))
    rel_afectiva.append(std.mean(aux_a))
    
    
```

```python
# Creamos el dataframe de opinion sobre uso de internet según relación

df2_satisfaccion = pd.DataFrame()

df2_satisfaccion["familia"] = rel_familia
df2_satisfaccion["social"] = rel_social
df2_satisfaccion["afectiva"] = rel_afectiva

df2_satisfaccion
```

# Pregunta 3 

**¿Cómo son de satisfactorias las relaciones afectivas según el nivel de estudios de cada 
componente de la pareja? ¿y según a su situación laboral?**

```python
df3 = df[ ( df["CLASESUB"] <= 5 ) & ( df["ESTUDIOS"] <= 6 ) ]
```

```python
clase = ["alta y media alta", "media-media", "media-baja", "trabajadora/obrera/ploretario", "baja/pobre"]
estudios = ["sin estudios", "primaria", "secundaria", "bachillerato", "FP", "superiores"]

clase_i = []
estudios_i = []
```

```python
for i, x in enumerate(df3["CLASESUB"]):
    
    clase_i.append(clase[x-1])
    
for j, y in enumerate(df3["ESTUDIOS"]):
    
    estudios_i.append(estudios[y-1])
```

```python
# Creamos el dataframe 

df3_estudios_clasesocial = pd.DataFrame()

df3_estudios_clasesocial["genero"] = df3["SEXO_NOMBRE"]
df3_estudios_clasesocial["edad"] = df3["EDAD_RANGO"]
df3_estudios_clasesocial["afectiva"] = df3["P1_3"]
df3_estudios_clasesocial["clase_social"] = clase_i
df3_estudios_clasesocial["nivel_estudios"] = estudios_i
```

```python
df3_estudios_clasesocial = df3_estudios_clasesocial[df3_estudios_clasesocial["afectiva"] < 11]
```
